### Statistical Modeling

```python
# Scatter plot of the active power vs. wind speed
df.plot.scatter(y='ActivePower', x='WindSpeed')
plt.title('Power output vs. wind speed', fontsize=12) 
```

Here we can see that the turbine can generate power when the wind speed is 2.5m/s, and the maximum output power is about 8m/s. Although there are some noise of the data, it is a exponential distribution. So we will create a exponential funtion to capture the trend.

```python
# The raw statistical model to capture data trend
X = np.arange(0, 20, 0.1)
Y = 1725 / (1.0 + np.exp(0.9*(-X+5)))
plt.plot(X,Y) 
plt.show()
```

Now, we will define a function to capture the trend and use curve_fit from scipy to find the optimal parameters of the function.

```python
# Define a funcation
def stat_func(X,c1,c2):
    return 1725 / (1.0 + np.exp(c1*(-X+c2)))
```

```python
# Use curve_fit to find the best parameters
c,cov = curve_fit(stat_func,df['WindSpeed'],df['ActivePower'], p0=[0.9, 5], maxfev=1000)
print(c)
```

Now we find the best parameters to capture the trend of the output power.

```python
# define and empty array to put the power predictions into
n = len(df['WindSpeed'])

predicted_power = np.empty(n)
for i in range(n):
    predicted_power[i] = graph_func(df['WindSpeed'][i],c[0],c[1]) # use values calculated in the curve_fit equation above
    
plt.plot(df['WindSpeed'],predicted_power,'r.')
plt.plot(df['WindSpeed'],df['ActivePower'],'b.', alpha=0.01 )
plt.ylabel('Output Power | (kW)', fontsize=12)
plt.xlabel('Windspeed m/s', fontsize=12)
plt.title('Power output vs. wind speed', fontsize=12 )
plt.legend(['Predicted','Actual'], fontsize=12)
```

```python
print('R-squared = {:.2f}'.format(r2_score(df['ActivePower'],predicted_power)))
print('The mean absolute error (MAE) = {:.2f}'.format(mean_absolute_error(df['ActivePower'],predicted_power)))
print('The RMSE error (RMSE) = {:.2f}'.format(mean_squared_error(df['ActivePower'],predicted_power, squared=False)))
```

Here, we can see the R-squre for this method is really high because our statistical model can perfectly capture the power output trend. Althogh it can predict the power output based on the trend, but it cannot do the seasonal analysis. Therefore, we introduce seasonal ARIMA (SARIMA) to do the prediction.

First, we use Dicker-Fuller test to check the stationality of the data and we find the p value is smaller than 0.05.

### Preprocessing 

```python
# Dicker-Fuller test
dftest = adfuller(df.ActivePower, autolag = 'AIC')
print(" ADF : ",dftest[0])
print(" p-Value : ", dftest[1])
print(" Num Of Lags : ", dftest[2])
print(" Num Of Observations Used For ADF Regression and Critical Values Calculation :", dftest[3])
print(" Critical Values :")
for key, val in dftest[4].items():
    print("\t",key, ": ", val)
```

```python
# Resample the data per day becasue we want to see the seasonal data per day
df_daily  = df[['ActivePower','WindSpeed']].resample('D').mean()
df_daily.dropna(inplace=True) 
```

```python
# Check the seasonal nature of the data set using the seasonal decompose function and setting the period equal to 31 (a month)
ss_check = seasonal_decompose(df_daily['ActivePower'].dropna(), period=31, model='additive',extrapolate_trend=True)
ss_check.plot()
plt.show()
```

We can see the data is seasonal, so we are sure we can use SARIMA to do the prediction.

```python
model1=pm.auto_arima(df_daily['ActivePower'],start_p=0,d=0,start_q=0,
          max_p=15,max_q=5, 
          start_P=0,D=0, start_Q=0, 
          max_P=8, max_Q=8, 
          m=12, seasonal=True, stationary=True,
          error_action='warn',trace=True,
          supress_warnings=True,stepwise=True,
          random_state=20,n_fits=100)
```

```python
model1.summary()
```

```python
# Prepare the train and test data sets. Here, we use 80% of the data as train.
X = df_daily['WindSpeed']
y = df_daily['ActivePower']
X_train, X_test, y_train, y_test = train_test_split(X,y ,
                                   random_state=104, 
                                   test_size=0.20, 
                                   shuffle=True)
```

### Modeling using SARIMA

```python
# Set up the SARIMA model and train the model
p = 6
d = 0
q = 0
P = 3
D = 0
Q = 0
m = 12

model = SARIMAX(y_train, order=(p,d,q), seasonal_order=(P,D,Q,m), dynamic=True)
model_fit = model.fit(disp=0) 
```

```python
# Make a prediction
test_pred = model_fit.predict(start=1, end=len(y_test))
```

```python
# make sure they have the same index
test_pred.index = y_test.index
```

```python
df_daily_pred = pd.DataFrame(data={'Predictions':test_pred, 'Actuals':y_test })  
df_daily_pred.sort_index(inplace=True)
df_daily_pred.head()
```

```python
print('(R-square) = {:.2f}'.format(r2_score(df_daily_pred['Actuals'],df_daily_pred['Predictions'])))
print('The mean absolute error (MAE) = {:.2f}'.format(mean_absolute_error(df_daily_pred['Actuals'],df_daily_pred['Predictions'])))
print('The RMSE error (RMSE) = {:.2f}'.format(mean_squared_error(df_daily_pred['Actuals'],df_daily_pred['Predictions'], squared=False)))
print('The Mean absolute percentage error (MAPE)  = {:.2f}'.format(mean_absolute_percentage_error(df_daily_pred['Actuals'],df_daily_pred['Predictions'])))   
```

Here, we can see the R-squire is low, so we try XGBoost model.

### Modeling using XGBoost

```python
model2 = xgb.XGBRegressor(n_estimators=1000)
```

```python
model2.fit(X_train, y_train, eval_set=[(X_train, y_train), (X_test, y_test)],
          early_stopping_rounds=50, #stop if 50 consequent rounds without decrease of error
          verbose=False) 
```

```python
test_pred2 = model2.predict(X_test)
```

```python
df_daily_pred2 = pd.DataFrame(data={'Predictions':test_pred2, 'Actuals':y_test })  
df_daily_pred2.sort_index(inplace=True)
df_daily_pred2.head()
```

```python
print('(R-square) = {:.2f}'.format(r2_score(df_daily_pred2['Actuals'],df_daily_pred2['Predictions'])))
print('The mean absolute error (MAE) = {:.2f}'.format(mean_absolute_error(df_daily_pred2['Actuals'],df_daily_pred2['Predictions'])))
print('The RMSE error (RMSE) = {:.2f}'.format(mean_squared_error(df_daily_pred2['Actuals'],df_daily_pred2['Predictions'], squared=False)))
print('The Mean absolute percentage error (MAPE)  = {:.2f}'.format(mean_absolute_percentage_error(df_daily_pred2['Actuals'],df_daily_pred2['Predictions'])))   
```

```python
plt.plot(df_daily_pred2['Predictions'],label='Prediction')
plt.plot(df_daily_pred2['Actuals'], color='red',label='Actual')

plt.ylabel('Output Power')
plt.legend()
plt.tight_layout()
plt.grid()
plt.show()
plt.rcParams["figure.figsize"] = (40,3)
```

Here, we can see that R squre is 0.97, which is pretty good. This means that XGBoost model works well for this output power prediction.

Here, we don't use GridSearch CV is because it dosen't work well for the time series data. And our current model works well without using GridSearch CV.
